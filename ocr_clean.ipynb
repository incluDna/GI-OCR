```python
import pandas as pd
import re
```

```python
df = pd.read_excel("ocr_output5.xlsx")

df.head()
```

```python
def clean_branch(text):
    text = " ".join(str(text).split())

    idx = text.rfind("ขา")
    idx2 = text.rfind("ยา")

    idx = max(idx, idx2)

    if idx != -1:
        result = text[idx + 2:].strip()
        result = result.lstrip(":- ")
        return result

    return text


df["ชื่อสาขา"] = (
    df["ชื่อสาขา"]
    .fillna("")
    .apply(clean_branch)
)

df[["ชื่อสาขา"]].head(20)
```

```python
df.to_excel(
    "ocr_output_clean.xlsx",
    index=False
)

print("saved")
```

แก้เลขเอกสาร มี 12 หลัก มักขึ้นต้นด้วย เลข 5

```python
import pandas as pd
import os
import re

INPUT_FILE = "ocr_output_clean.xlsx"

base, ext = os.path.splitext(INPUT_FILE)
OUTPUT_FILE = f"{base}_update{ext}"

df = pd.read_excel(INPUT_FILE)

def clean_doc_no(value):
    # เอาเฉพาะตัวเลข
    digits = re.sub(r"\D", "", str(value))

    # หาเลข 12 หลักที่ขึ้นต้นด้วย 5
    match = re.search(r"5\d{11}", digits)

    if match:
        return match.group(0)

    return ""

df["เลขที่เอกสาร"] = df["เลขที่เอกสาร"].apply(clean_doc_no)

df.to_excel(OUTPUT_FILE, index=False)

print("saved:", OUTPUT_FILE)
```
